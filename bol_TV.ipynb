```python
import requests
from bs4 import BeautifulSoup
```

```python
result=requests.get("https://www.bol.com/nl/l/tv-s/N/7291/filter_N/4278212706+4278212668+4278212630+4276191829+4275059034+4274989803/?promo=televisies_360__A_1290-63717-grote-tv%27s_2_")
print(result.status_code)
```

```python
src=result.content
soup=BeautifulSoup(src,'lxml')
print(soup)
```

```python

isim=soup.find("ul",{"id":"js_items_content"}).find("li",{"data-id":"9200000114242718"}).find("a",{"class":"product-title px_list_page_product_click"})
print(isim.get_text())
```

```python
fiyat=soup.find("ul",{"id":"js_items_content"}).find("li",{"data-id":"9200000114242718"}).find("span",{"class":"promo-price"})

fiyat=fiyat.get_text().split(" ")
print(fiyat[0])
```

```python
liste=[]
listevericek=soup.find_all("li",{"class":"product-item--row js_item_root"})
i=0
for tekveri in listevericek:
    i=i+1
    print(tekveri.attrs["data-id"])
    liste.append(tekveri.attrs["data-id"])
print("total urun : "+str(i))
```

```python
def isimcek(id):
    
    
    isim=soup.find("ul",{"id":"js_items_content"}).find("li",{"data-id":id}).find("a",{"class":"product-title px_list_page_product_click"})
    print(isim.get_text())
    return isim.get_text()
```

```python
def fiyatcek(id):
    fiyat=soup.find("ul",{"id":"js_items_content"}).find("li",{"data-id":id}).find("span",{"class":"promo-price"})

    fiyat=fiyat.get_text().split(" ")
    print(fiyat[0])
    return fiyat[0]
    
```

```python
def idtopla():
    liste=[]
    listevericek=soup.find_all("li",{"class":"product-item--row js_item_root"})
    i=0
    for tekveri in listevericek:
        i=i+1
        liste.append(tekveri.attrs["data-id"])
    print(str(i))
   
```

```python

def cek():
    idtopla()
    for tekveri in liste:
        isimcek(tekveri)
        fiyatcek(tekveri)
        print(isimcek(tekveri), fiyatcek(tekveri))
```

```python
cek()
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
result=requests.get("https://www.bol.com/nl/l/boeken/N/8299/?bltgh=iGzBS0nTjARKK2axvadCYw.2_3_4.5.CategoryImage")
print(result.status_code)
```

```python
src=result.content
soup=BeautifulSoup(src,'lxml')
print(soup)
```

```python
def like(id):
    liste_deneme=soup.find("li",{"data-id":id}).find("div",{"class":"star-rating"})
    a=liste_deneme.attrs["title"].split(" ")
    return a[1]
    
```

```python
def info(id):
    liste_deneme=soup.find("li",{"data-id":id}).find("ul",{"class":"product-small-specs"})
    for a in liste_deneme:
        return liste_deneme.get_text()
    
    
```

```python
def aciklama(id):
    liste_deneme=soup.find("li",{"data-id":id},recursive=True).find("p",{"class":"medium--is-visible"})
    return liste_deneme.get_text().strip()
    
```

```python
def yazarcek(id):
    liste_deneme=soup.find("li",{"data-id":id}).find("a",{"itemprop":"name"})
    return liste_deneme.get_text()
    
```

```python
def paracek(id):
    
    liste_deneme=soup.find("li",{"data-id":id}).find("div",{"class":"price-block__highlight"}).find("meta")
    return liste_deneme.attrs["content"]
```

```python
def kitapismi(id):
    liste_deneme=soup.find("li",{"data-id":id}).find("div",{"class":"product-title--inline"}).find("a")

    return liste_deneme.get_text()

    
```

```python
def resimcek(id):

    try:
        liste_deneme=soup.find("li",{"data-id":id}).find("div",{"class":"h-o-hidden"}).find("img").attrs["data-src"]
  
        
    except:
        liste_deneme=None
        
        
        
    
    return liste_deneme

    
```

```python
def idtopla():
    idlist=[]
    aa=soup.findAll("li",{"class":"product-item--row js_item_root"})
   
    for b in aa:
        idlist.append(b.attrs["data-id"])
    return idlist
```

```python
for idler in idtopla():
    print(resimcek(idler))
```

```python
#add  pip install PTable
from prettytable import PrettyTable
table =PrettyTable(["isim","yazar","fiyat","like"])
i=0
while 1:
    i=i+1
    try:
        
        result=requests.get("https://www.bol.com/nl/l/boeken/N/8299/?page="+str(i))
        src=result.content
        soup=BeautifulSoup(src,'lxml')
        for idler in idtopla():
            table.add_row([kitapismi(idler),yazarcek(idler),paracek(idler),like(idler)])
    except:
        print("toplam sayfa : "+str(i))
        break
        
   
```
